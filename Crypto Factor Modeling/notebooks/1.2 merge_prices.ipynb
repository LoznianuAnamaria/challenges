```python
import pandas as pd
```

```python
input_train_file_path = '../data/train_targets.parquet'
input_live_file_path = '../data/live_targets.parquet'

input_historical_data_file_path = '../data/historical_data.csv'
output_merged_file_path = '../data/merged_historical_prices.parquet'
```

```python
train_df = pd.read_parquet(input_train_file_path)
live_df = pd.read_parquet(input_live_file_path)

historical_data = pd.read_csv(input_historical_data_file_path)
```

```python
# Merge the data on 'symbol' and 'date'
merged_data = pd.merge(train_df, historical_data, on=['symbol', 'date'], how='left')

# Save the merged data
merged_data.to_parquet(output_merged_file_path, index=False)
merged_data.head(10)
```
