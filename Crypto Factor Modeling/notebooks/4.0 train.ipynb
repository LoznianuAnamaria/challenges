```python
import lightgbm as lgb
import pandas as pd
from numerapi import NumerAPI
from typing import List

import joblib
```

```python
input_file_path = '../data/merged_historical_prices.parquet'

output_model_path = '../data/crypto_model.pkl'
```

```python

# Load the merged dataset
data = pd.read_parquet(input_file_path)

# Generate training features
def generate_training_features(df: pd.DataFrame) -> List[str]:
    return ['open', 'high', 'low', 'close', 'volume']

# Add training features for each (symbol, date)
feature_cols = generate_training_features(data)

# Initialize the LightGBM model
model = lgb.LGBMRegressor(
    n_estimators=2000,
    learning_rate=0.01,
    max_depth=5,
    num_leaves=2 ** 5,
    colsample_bytree=0.1
)

# Train the model
model.fit(
    data[feature_cols],
    data["target"]
)

# Save the model
joblib.dump(model, output_model_path)
```
