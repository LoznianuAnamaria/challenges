```python
input_parquet_file_path = '../data/interm/1.0_taxdata.parquet'
```

# Overview of the dataset

```python
# Read the parquet file into a DataFrame
df = pd.read_parquet(input_parquet_file_path)

# Print the number of rows and columns
print(f'Number of rows: {df.shape[0]}')
print(f'Number of columns: {df.shape[1]}')
```

```python
# Display summary statistics
df.describe()
```

```python
df.head()
```

# Distribution of data across years

```python
# Group by 'ANNEE' and count the number of rows for each year
yearly_distribution = df.groupby('ANNEE').size()

# Plot the distribution
plt.figure(figsize=(12, 6))
yearly_distribution.plot(kind='bar', color='skyblue')
plt.title('Tax data distribution across Years')
plt.xlabel('Year')
plt.ylabel('Tax Data')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Dataset completeness info

```python
# Calculate the percentage of missing values per column
missing_percentage = df.isnull().mean() * 100

# Calculate the overall fill percentage
overall_fill_percentage = 100 - missing_percentage.mean()

# Calculate the number of columns with 0% missing values
columns_with_no_missing = (missing_percentage == 0).sum()

# Define a threshold for high missing values (e.g., more than 70% missing)
high_missing_threshold = 70
columns_with_high_missing = (missing_percentage > high_missing_threshold).sum()

print(f'Overall fill percentage: {overall_fill_percentage:.2f}%')
print(f'Number of empty columns: {columns_with_no_missing}')
print(f'Number of columns with more than {high_missing_threshold}% missing values: {columns_with_high_missing}')

# Plot the distribution of missing values
plt.figure(figsize=(12, 6))
missing_percentage.plot(kind='hist', bins=50, color='skyblue')
plt.title('Distribution of Missing Values Across Columns')
plt.xlabel('Percentage of Missing Values (%)')
plt.ylabel('Number of Columns')
plt.grid(axis='y')

# Customize the x and y axis ticks for better readability
plt.xticks(ticks=range(0, 101, 10), labels=[f'{i}%' for i in range(0, 101, 10)])
plt.yticks(ticks=range(0, int(missing_percentage.max()) + 1, 5))

plt.tight_layout()
plt.show()
```

# Fill in rate of specific important columns

```python
columns_to_check = ['code INSEE', 'COMMUNE', 'Libellé commune', 'DEPARTEMENT', 'LIBELLE DEPARTEMENT']
fill_rates = df[columns_to_check].notnull().mean() * 100

# Print the fill rates for the specified columns
for column, fill_rate in fill_rates.items():
    print(f'Fill rate for {column}: {fill_rate:.2f}%')


# Plot the fill rates
colors = ['red' if i == 0 or i == len(columns_to_check) -1 else 'skyblue' for i in range(len(fill_rates))]

plt.figure(figsize=(10, 6))
fill_rates.plot(kind='bar', color=colors)
plt.title('Fill Rates for Specific Columns')
plt.xlabel('Columns')
plt.ylabel('Fill Rate (%)')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage of non-missing values for each column
non_missing_percentage = df.notna().mean() * 100

# Create a DataFrame for better visualization
filled_percentage_df = pd.DataFrame(non_missing_percentage, columns=['Percentage Filled'])

# Print the DataFrame (optional)
print(filled_percentage_df)
```
