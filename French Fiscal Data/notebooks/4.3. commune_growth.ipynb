```python
%pip install matplotlib
%pip install pandas

import gc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_parquet_file_path = '../data/interm/4.0_commune.parquet'
```

```python
code_insee_identifier = 'Clean code INSEE'

data_df = pd.read_parquet(input_parquet_file_path)
tax_columns = data_df.columns[8:]
```

```python
# Ensure all relevant columns are treated as numeric
data_df[tax_columns] = data_df[tax_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the total tax revenue for each row
data_df['Total_Taxes'] = data_df[tax_columns].sum(axis=1)

# Function to calculate growth rate
def calculate_growth_rate(start, end):
    if start == 0:
        return None  # Avoid division by zero
    return ((end - start) / start) * 100

# Define time periods
time_periods = [5, 10, 15, 20]

# Add the 'LABEL' column to the growth_rates DataFrame
growth_rates_list = []

# Calculate growth rates for each municipality
for municipality in data_df['Clean code INSEE'].unique():
    row = {'Municipality': municipality}
    municipality_data = data_df[data_df['Clean code INSEE'] == municipality].sort_values(by='ANNEE')
    
    for period in time_periods:
        start_year = municipality_data['ANNEE'].max() - period
        if start_year in municipality_data['ANNEE'].values:
            start_revenue = municipality_data[municipality_data['ANNEE'] == start_year]['Total_Taxes'].values[0]
            end_revenue = municipality_data[municipality_data['ANNEE'] == municipality_data['ANNEE'].max()]['Total_Taxes'].values[0]
            # Make sure that the communes with a 0 starting || ending points are eliminated from the calculations
            if start_revenue >= 100 and end_revenue >= 100:
                row[f'{period}_year'] = calculate_growth_rate(start_revenue, end_revenue)
        else:
            row[f'{period}_year'] = None
    
    row['LABEL'] = municipality_data['LABEL'].values[0]  # Add the label
    growth_rates_list.append(row)

growth_rates = pd.DataFrame(growth_rates_list)

# Update the plot functions to use 'LABEL' for the x-axis
def plot_top_growth_rates(top_municipalities, period):
    top_df = top_municipalities[period].dropna()
    plt.figure(figsize=(12, 6))
    plt.bar(top_df['LABEL'], top_df[f'{period}_year'], color='skyblue')
    plt.title(f'Top 10 Growing Municipalities over {period} Years')
    plt.xlabel('Municipality')
    plt.ylabel('Growth Rate (%)')
    plt.yscale('log')  # Use logarithmic scale
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()  # Adjust layout to ensure labels fit
    plt.grid(True, which="both", ls="--")
    plt.show()
    
    # Print the results
    print(top_df[['Municipality', 'LABEL', f'{period}_year']])

def plot_bottom_growth_rates(bottom_municipalities, period):
    bottom_df = bottom_municipalities[period].dropna()
    plt.figure(figsize=(12, 6))
    plt.bar(bottom_df['LABEL'], bottom_df[f'{period}_year'], color='salmon')
    plt.title(f'Top 10 Municipalities with Downward Trend over {period} Years')
    plt.xlabel('Municipality')
    plt.ylabel('Growth Rate (%)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()  # Adjust layout to ensure labels fit
    plt.grid(True)
    plt.show()
    
    # Print the results
    print(bottom_df[['Municipality', 'LABEL', f'{period}_year']])

# Find the top 10 municipalities by growth rate for each period
top_municipalities = {}
for period in time_periods:
    top_municipalities[period] = growth_rates[['Municipality', 'LABEL', f'{period}_year']].sort_values(by=f'{period}_year', ascending=False).head(10)

# Find the top 10 municipalities by downward growth rate for each period
bottom_municipalities = {}
for period in time_periods:
    bottom_municipalities[period] = growth_rates[['Municipality', 'LABEL', f'{period}_year']].sort_values(by=f'{period}_year').head(10)

# Plot the results
for period in time_periods:
    plot_top_growth_rates(top_municipalities, period)
    plot_bottom_growth_rates(bottom_municipalities, period)
```
